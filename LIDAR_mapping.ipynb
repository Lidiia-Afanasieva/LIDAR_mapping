```python
import numpy as np
```

```python
txt = open('/content/examp6.txt')
data = []

for line in txt:
    # print(line)
    data.append(line[line.find("; ") + 2:-1].split(', '))

data = np.array(data, dtype=np.float64)
```

```python
data.shape
```

```python
alf = 0.006 # rads

def angle_calc(data):
    map = np.zeros([121, 121], dtype=str)
    map[56][56] = '9'

    # for line in range(len(data)):
    #     # 1st
    #     # print(len(data))
    #     # print(line)
    for n, rad in enumerate(data[:83]):
        # print(rad)
        rad = round(rad, 3) * 10
        angle = (n+1) * alf
        y = int(round(np.sin(angle) * rad))
        x = int(round(np.cos(angle) * rad))
        map[y+56][x+56] = 'Ж'

    # 2nd
    for n, rad in enumerate(data[84:340]):
        rad = round(rad, 3) * 10
        angle = (n+1) * alf
        y = int(round(- np.sin(angle) * rad))
        x = int(round(np.cos(angle) * rad))
        map[y+56][x+56] = 'Ж'

    # 3rd
    for n, rad in enumerate(data[341:597]):
        rad = round(rad, 3) * 10
        angle = (n+1) * alf
        y = int(round(- np.cos(angle) * rad))
        x = int(round(- np.sin(angle) * rad))
        map[y+56][x+56] = 'Ж'

    # 4th
    for n, rad in enumerate(data[598:]):
        rad = round(rad, 3) * 10
        angle = (n+1) * alf
        y = int(round(np.sin(angle) * rad))
        x = int(round(- np.cos(angle) * rad))
        map[y+56][x+56] = 'Ж'

    for i in range(len(map)):
        for j in range(len(map[i])):
            if map[i][j] != 'Ж': map[i][j] = ' '

    map[56][56] = '9'
    for i in range(2):
        map[56+1][56] = '9'
        map[56-1][56] = '9'
        map[56][56+1] = '9'
        map[56][56-1] = '9'

    return map

'''
    округлить то сотней
    добавить промежуточные
'''
```
